# NOTES:
- MOST the pandas dataframe functionality will work on pyspark data frames

- to see dataframe in tabular fashion: df.show()

- Select a particular(one) column:  df.select('columnName')
- select multiple columns df.select(['col1','col2'])
- # Note that splicing wont work

- see data types df.dtypes()

- describe: (same as pandas) df.describe()

# Note both add and drop are Not inplace operation, we need to assign it to a variable 
# eg df = df.drop('colName')
- Adding/Update columns: df.withColumn('NewColumn')
- remove column: df.drop('colName')
- rename_column: df.withColumnRenamed('NewName', name)


# Part 2: Handling missing values
- drop entire row if any null/na in column : df.na.drop()
- na.drop has 3 features how, thresh and subset
- na.drop(how='all'): only drop row if all col value is null
- na.drop(how='any', thres=2) : only drop row if atleast 2 null/na value are pret
- na.drop(how='any', subset=['Name']) : only drop row if any values inside the subset param (Name column) is null

- filling missing Value  with certain value :df.na.fill( certainValue,[columnNamesn where we replace])
- fill the null value with mean : Do using imputer function

# Filter values
- filter values and search certain columns : df.filter('Salary<=200').select(['col1','col2'])


# Groupby and aggregate. They always work together
'- df.groupBy('Country').sum()
